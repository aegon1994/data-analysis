```python
import requests
from bs4 import BeautifulSoup
url = 'https://movies.yahoo.com.tw/movie_thisweek.html'
r = requests.get(url)
response = r.text
d = BeautifulSoup(r.text,'html5lib')
print(d)
```

```python
cname = d.find_all('div', class_ ="release_info")
for name in cname:
    ctitle = name.find('div',class_="release_movie_name").a.text.strip()
    etitle = name.find('div', class_="en").text.strip()
    date = name.find('div', class_="release_movie_time").text.strip()
    j = name.find('div',class_="level_name").text
    jr = name.find('div',class_="leveltext").span.text.strip()
    print(ctitle)
    print(etitle)
    print(date)
    print(j," ",jr,"\n")
```
